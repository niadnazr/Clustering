```python
import pandas as pd
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
crime.drop(['Unnamed: 0'],axis = 1,inplace = True)
```

```python
crime = crime.rename({'Murder':'MURDER','Assault':'ASSAULT','UrbanPop':'URBAN_POP','Rape':'RAPE'},axis=1)
crime
```

```python
crime.info()
```

HIERARCHIAL CLUSTERING

```python
# NORMALIZING FUNCTIONS
def norm_func(i):
    x = (i - i.min())/(i.max()-i.min())
    return (x)
```

```python
#normalized data frame
df_norm = norm_func(crime)
```

```python
#creating dendogram
dendogram = sch.dendrogram(sch.linkage(df_norm,method='single'))
```

```python
hc = AgglomerativeClustering(n_clusters = 4, affinity = 'euclidean', linkage = 'single')
```

```python
# save clusters for chart
y_hc = hc.fit_predict(df_norm)
y_hc
```

```python
crime['H_CLUSTERID'] = hc.labels_
crime
```

K MEANS

```python
#THE K-MEANS ALGORITHM AIMS TO CHOOSE THE CENTROIDS THAT MINIMISE THE INERTIA OR WITHIN-CLUSTER SUM OF SQUARESCRITERION:
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i,random_state = 0)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11),wcss)
plt.title('ELBOW METHOD')
plt.xlabel('NUMBER OF CLUSTERS')
plt.ylabel('WCSS')
plt.show()
```

```python
#BUILDING CLUSER ALGORITHM
clusters_new = KMeans(4,random_state = 42)
clusters_new.fit(df_norm)
```

```python
clusters_new.labels_
```

```python
#assigning clusters to the data set
crime['CLUSTERID_NEW'] = clusters_new.labels_
```

```python
#STANDARDIZED VALUES ARE
clusters_new.cluster_centers_
```

```python
crime.groupby('CLUSTERID_NEW').agg(['mean']).reset_index()
```

DB SCAN

```python
#normalizing heterogenious numerical data using scalar fit transfrom to dataset.
from sklearn.preprocessing import StandardScaler
crime_norm = StandardScaler().fit_transform(crime)
crime_norm
```

```python
#DBSCAN CLUSTERING
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=1,min_samples = 4)
dbscan.fit(crime_norm)
```

```python
#noisy samples are given in label -1
dbscan.labels_
```

```python
crime.groupby('CLUSTERS').agg(['mean']).reset_index()
```

```python
crime['CLUSTERS'] = dbscan.labels_
crime
```
